```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.myfootballfacts.com/premier-league-winners/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
```

```python
table.find_all('th')
```

```python
PremTitles = table.find_all('th')
```

```python
PremTitles
```

```python
Titles = [title.text.strip() for title in PremTitles]

print(Titles)
```

```python
columnData = table.find_all('tr')
```

```python
for row in columnData[2:]:
    rowData = row.find_all('td')
    individ = [data.text.strip() for data in rowData]
    print(individ)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns = Titles)
```

```python
df
```

```python
df.drop(columns = '')
```

```python
df = df.drop(columns = '')
```

```python
for row in columnData[2:]:
    rowData = row.find_all('td')
    individ = [data.text.strip() for data in rowData]
    length = len(df)
    df.loc[length] = individ
```
